```python
#a function to split a column, so that each column has unique data in it

def split_column(df,column_name):
    #split by comma
    n = df[column_name].str.split(",",expand = True)
    #merge
    n1 = df.merge(n,left_index=True,right_index =True)
    #delete the duplicate column
    n2 = n1.drop(column_name,1)
    
    
    #get length of created new columns
    count = [i for i in range(len(n.columns))]
    
    #rename each new column the original name plus its number
    for i in count:
        n2.rename(columns = {count[i]:(column_name + (str(i+1)))}, inplace = True)
    df = n2
    return df
    
    

test1 = split_column(test,"knownForTitles")
test1
```
